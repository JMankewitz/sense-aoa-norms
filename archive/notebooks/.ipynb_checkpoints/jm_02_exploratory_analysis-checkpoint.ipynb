```python
def compareEstimations(raw_estimation_df, lowest=False, cdi_lex_class_list=[]):
    raw_estimation_df['word'] = raw_estimation_df.sense_question.apply(lambda question: question.split(' ')[0])
    raw_estimation_df['lemma'] = raw_estimation_df.word.apply(lambda word: lemmatizer.lemmatize(word).strip())

    producing_estimations = raw_estimation_df[raw_estimation_df['question_group']=='say']
    mean_producing = producing_estimations.groupby(by=['lemma'])[['aoa_estimation']].agg('mean').reset_index()

    if lowest == True:
        words = []
        min_estimations = []
        mean_estimations = producing_estimations.groupby(by=['lemma','sense_question'])[['aoa_estimation']].agg('mean').reset_index()

        for lemma in list(set(mean_estimations.lemma)):
            min_estimation = min(list(mean_estimations[mean_estimations['lemma']==lemma].aoa_estimation))
            words.append(lemma)
            min_estimations.append(min_estimation)
    
        mean_producing = pd.DataFrame(data={'lemma':words, 'min_aoa_estimation':min_estimations})

    wordbank_aoa['wordbank_aoa_in_years'] = wordbank_aoa.aoa.apply(lambda aoa: aoa/365)
    if cdi_lex_class_list != []:
        estimations_vs_others = pd.merge(mean_producing,
                                 wordbank_aoa[wordbank_aoa['lexical_class'].isin(cdi_lex_class_list)][['wordbank_aoa_in_years','lemma']],
                                left_on='lemma', right_on='lemma',
                                how='inner')
    else:
        estimations_vs_others = pd.merge(mean_producing,
                                 wordbank_aoa[['wordbank_aoa_in_years','lemma']],
                                left_on='lemma', right_on='lemma',
                                how='inner')

    estimations_vs_others = pd.merge(estimations_vs_others,
                                    kuperman_aoa[['Word','Rating.Mean']],
                                    left_on='lemma',
                                    right_on='Word',
                                    how='inner')
    return estimations_vs_others
```

```python
import re
import string
lemmatizer = WordNetLemmatizer()

def matchToWordnetSynset(survey_sense, question_dict=question_dict, verbose=False):
    if survey_sense not in question_dict.keys():
        print(survey_sense, "not in question dict")
        return None
    match_definition = re.search('(?<=\-).*?(?=\( E)', question_dict[survey_sense])
    match_definition = str(match_definition.group()).strip()
    match_definition = match_definition.strip(string.punctuation.replace('()', '')).replace('"', '').replace("`","'")

 
    match_gloss = survey_sense.split(" ")[0]
    match_lemma = lemmatizer.lemmatize(match_gloss)

    #word_synsets = wn.synsets(match_lemma, match_pos)
    word_synsets = wn.synsets(match_lemma)

    no_match = True
    bad_def = ''
    for possible_synset in word_synsets:
        poss_definition = possible_synset.definition().replace("`", "'")
        poss_definition_2 = re.split("[" + ';,'+ "]+",poss_definition)[0]
        poss_definition_2 = poss_definition.replace("`", "'")
        try_match_2 = re.search('(?<=).*?(?=\; ;)',poss_definition_2)
        if try_match_2:
            try_match_2 = str(try_match_2.group()).strip()
        else:  
            try_match_2 = re.search('(?<=).*?(?=\; \-)',poss_definition_2)
            if try_match_2:
                try_match_2 = str(try_match_2.group()).strip()
        if match_definition == poss_definition:
            return possible_synset.name()
        elif match_definition == poss_definition_2:
            return possible_synset.name()
        elif match_definition.split(',')[0] == poss_definition_2.replace(',',''):
            return possible_synset.name()
        elif match_definition == poss_definition_2.replace('"', '') or match_definition == poss_definition.replace('"', ''):
            return possible_synset.name()
        elif match_definition.split(";")[0] == poss_definition_2 or match_definition.split(";")[0] == poss_definition:
            return possible_synset.name()
        elif match_definition == poss_definition.strip(string.punctuation):
            return possible_synset.name()
        elif match_definition == poss_definition.replace(".", ''):
            return possible_synset.name()
        elif match_definition == poss_definition_2:
            return possible_synset.name()
        elif match_definition.replace(";", '') == try_match_2:
            return possible_synset.name()
        elif match_definition == try_match_2:
            return possible_synset.name()
        else:
            if survey_sense == 'circus n_say_1' and verbose:
                print("source def:", match_definition)
                print(poss_definition_2)
            bad_def = match_definition
            
    #if doesnt match lemma, try gloss
    for possible_synset in wn.synsets(match_gloss):
        poss_definition = possible_synset.definition()
        poss_definition = re.split("[" + ';,'+ "]+",poss_definition)[0]
        poss_definition = poss_definition.replace("`", "'")
        if match_definition == possible_synset.definition():
            return possible_synset.name()
        elif match_definition == poss_definition:
            return possible_synset.name()
        else:
            bad_def = match_definition
            
    return "no_def_match"
```

```python
senses_to_match = clean_sense_survey[clean_sense_survey['question_type']!='attention_check']
senses_to_match = senses_to_match[['sense_question']]
senses_to_match = senses_to_match.groupby(by=['sense_question']).size().reset_index().drop(columns=[0])
senses_to_match
```

```python
senses_to_match['wordnet_sense'] = senses_to_match.sense_question.apply(lambda sense_q: matchToWordnetSynset(sense_q))
```

```python
print(len(senses_to_match[senses_to_match['wordnet_sense']=='no_def_match']))

def get_proposed_wordnet_sense(bad_sense_question):
    pos_map = {'adj':'a'}

    match_pos = re.search('(?<=\ ).*?(?=\_)', bad_sense_question)
    match_pos = str(match_pos.group()).strip()
    if match_pos in pos_map.keys():
        match_pos = pos_map[match_pos]
        
    match_sense_num = re.search('\d+(\.\d+)?$', bad_sense_question)
    match_sense_num = str(match_sense_num.group())
    
    if bad_sense_question not in question_dict.keys():
        print(bad_sense_question, "not in question dict")
        return None
    
    match_definition = re.search('(?<=\-).*?(?=\( E)', question_dict[bad_sense_question])
    match_definition = str(match_definition.group()).strip()
    match_definition = match_definition.strip(string.punctuation.replace('()', '')).replace('"', '').replace("`","'")
    match_gloss = bad_sense_question.split(" ")[0]
    match_lemma = lemmatizer.lemmatize(match_gloss)

    proposed_synset = str(match_lemma+'.'+match_pos+'.'+match_sense_num)
    return wn.synset(proposed_synset).name()

bad_sense_questions = senses_to_match[senses_to_match['wordnet_sense']=='no_def_match']
bad_sense_questions['wordnet_sense'] = bad_sense_questions.sense_question.apply(get_proposed_wordnet_sense)


converted_sense_questions_df = pd.concat([senses_to_match[senses_to_match['wordnet_sense']!='no_def_match'],
                                         bad_sense_questions])
print("all sense_questions converted:", set(converted_sense_questions_df.sense_question) == set(senses_to_match.sense_question))
converted_sense_questions_df
```

```python
def compareAdjustedEstimations(raw_estimation_df, lowest=False, cdi_lex_class_list=[]):
    raw_estimation_df['word'] = raw_estimation_df.sense_question.apply(lambda question: question.split(' ')[0])
    raw_estimation_df['lemma'] = raw_estimation_df.word.apply(lambda word: lemmatizer.lemmatize(word).strip())

    producing_estimations = raw_estimation_df[raw_estimation_df['question_group']=='say']
    mean_producing = producing_estimations.groupby(by=['lemma'])[['aoa_estimation']].agg('mean').reset_index()

    if lowest == True:
        words = []
        min_estimations = []
        mean_estimations = producing_estimations.groupby(by=['lemma','sense_question'])[['adjusted_estimations']].agg('mean').reset_index()

        for lemma in list(set(mean_estimations.lemma)):
            min_estimation = min(list(mean_estimations[mean_estimations['lemma']==lemma].adjusted_estimations))
            words.append(lemma)
            min_estimations.append(min_estimation)
    
        mean_producing = pd.DataFrame(data={'lemma':words, 'min_aoa_estimation':min_estimations})

    wordbank_aoa['wordbank_aoa_in_years'] = wordbank_aoa.aoa.apply(lambda aoa: aoa/365)
    if cdi_lex_class_list != []:
        estimations_vs_others = pd.merge(mean_producing,
                                 wordbank_aoa[wordbank_aoa['lexical_class'].isin(cdi_lex_class_list)][['wordbank_aoa_in_years','lemma']],
                                left_on='lemma', right_on='lemma',
                                how='inner')
    else:
        estimations_vs_others = pd.merge(mean_producing,
                                 wordbank_aoa[['wordbank_aoa_in_years','lemma']],
                                left_on='lemma', right_on='lemma',
                                how='inner')

    estimations_vs_others = pd.merge(estimations_vs_others,
                                    kuperman_aoa[['Word','Rating.Mean']],
                                    left_on='lemma',
                                    right_on='Word',
                                    how='inner')
    return estimations_vs_others
```
